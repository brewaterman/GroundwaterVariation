```python
#Example Dataframe
GWdata = {
    'dates': ["1/5/99","1/10/99","1/15/99","1/20/99","1/25/99","1/31/99","2/1/99","2/5/99","1/10/00"],
    'depths': [814.5, 814.33, 814.2, 814.11, 813.95, 814.05,814.21,814.26,814.29]
}
GWdata

df = pd.DataFrame(GWdata)
print(df)
df.dates = ["1/5/99","1/10/99","1/15/99","1/20/99","1/25/99","1/31/99","2/1/99","2/5/99", "1/10/00"]
df.depths = [814.5, 814.33, 814.2, 814.11, 813.95, 814.05,814.21,814.26, 814.29]
```

```python
#Variables and Define fx

def avgWell():
    """The goal of code is to calculate the average well depth for a specifc well. The
    input must be well data file (txt/csv) and the function will output a .csv file with the 
    3 columns: well name, average water level, and years of recorded data"""
           
    NumbDates = len(df.dates)
    sumdepth = sum(df.depths) 
    avgWD = sumdepth / NumbDates
    
    return avgWD

print(avgWell())
print(type(avgWell()))
```

```python
fig = plt.figure(figsize=(8, 6), dpi= 80)
plt.scatter(df.dates, df.depths, marker = "s", facecolors ="b") #fx to plot, x and y values
plt.ylabel("GW level from reference datum (ft)")
plt.xlabel("Dates")
plt.savefig('gwlevels.jpg')
plt.show()
```
